## CÓDIGO DE SOLUCIÓN ALUMNO 1

Se tomó como referencia los rendimientos de certificados de la tesorería (CETES) a un plazo de un año, siendo *10.87* el rendimiento

```python
#Instalamos nuestros paquetes y librerías

import pandas as pd
from datetime import date
import numpy as np
import numpy_financial as npf
```

```python
#Datos para la cuenta
tinteres = 0.1087
#Planeamos una fecha de término de la carrera en junio del 2025
año = 2.5
abono_año = 12
#Iniciamos en nuestra cuenta con 5000 pesos
inicial = 5000
meta = 85000
fecha_inicio = (date(2022,11,14))
```

```python
#calcular abonos mensuales
abono = -npf.pmt(tinteres/abono_año,año*abono_año,(meta-inicial))
abono
```

#### Debemos hacer abonos de 3,057 pesos

```python
#Empezamos a generar nuestro dataframe y los valores necesarios
rng = pd.date_range(fecha_inicio, periods=año*abono_año, freq="M")
rng.name = 'Fechas del abono'

#Columnas de abonos
df = pd.DataFrame(index = rng, columns=['Abono','Balance'],dtype='float')
#Fechas como índices
df.reset_index(inplace=True)

df.index+=1
df.index.name = 'Periodo'
df.head()
```

```python
#Definir abonos
df['Abono'] = abono
#Vector de los periodos
index_vector = np.arange(1,len(rng)+1)
#Balance de la cuenta objetivo
df['Balance'] = inicial * (1+tinteres/abono_año)**index_vector + \
                abono *                       (((1+tinteres/abono_año)**index_vector -1)/(tinteres/abono_año))


#Redondeo de cifras para mejor formato
df = df.round(2)
df
```
